```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
file_path = "CTG.xls"

df = pd.read_excel(file_path, sheet_name='Data', skiprows=1)
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Create a copy of the original DataFrame
df_cleaned = df.copy()

# 1. Drop the completely empty 'Unnamed' columns
empty_cols = ['Unnamed: 9', 'Unnamed: 31', 'Unnamed: 42', 'Unnamed: 44']
df_cleaned = df_cleaned.drop(columns=empty_cols)

# 2. Drop the redundant columns
redundant_cols = ['AC', 'FM', 'UC', 'DL', 'DS', 'DP']
df_cleaned = df_cleaned.drop(columns=redundant_cols)

# 3. Rename the '.1' columns
rename_map = {
    'AC.1': 'AC',
    'FM.1': 'FM',
    'UC.1': 'UC',
    'DL.1': 'DL',
    'DS.1': 'DS',
    'DP.1': 'DP'
}
df_cleaned = df_cleaned.rename(columns=rename_map)

# 4. Drop the useless 'DR' column (all zero)
df_cleaned = df_cleaned.drop(columns=['DR'])

# 5. Drop all rows that have any remaining missing values
df_cleaned = df_cleaned.dropna()
```

```python
# Verify the cleaning by checking the info and the first few rows
print("--- Info for the Cleaned DataFrame ---")
df_cleaned.info()

print("\n--- First 5 Rows of the Cleaned DataFrame ---")
df_cleaned.head()
```

```python
#Plotting of box graph

sns.set_style("whitegrid")

#create countplot of NSP column
plt.figure(figsize=(8, 5))
sns.countplot(x='NSP', data=df_cleaned)
plt.title('Distribution of Fetal Health Classes')
plt.xlabel('NSP Class (1=Normal, 2=Suspect, 3=Pathologic)')
plt.ylabel('Number of Samples')
plt.show()

```

### Analysis of target variable has significant class imbalance. Most classes are normal (1.0). This could be a challenge as our model could become biased towards the majority class. We need to make sure it can also identify 'suspect' and 'pathologic' classes accurately.

```python
# Create a boxplot for Abnormal Short-Term Variability (ASTV)
plt.figure(figsize=(10, 6))
sns.boxplot(x='NSP', y='ASTV', data=df_cleaned)
plt.title('Abnormal Short-Term Variability vs. Fetal Health Class')
plt.xlabel('NSP Class (1=Normal, 2=Suspect, 3=Pathologic)')
plt.ylabel('ASTV (%)')
plt.show()
```

### Boxplot reveals that having a higher ASTV value is a key characteristic of both 'Suspect' and 'Pathological' classes. Furthermore, the presence of many high-end outliers in these groups suggests that extreme levels of abnormal variability are a strong signal for fetal distress, making ASTV a critical feature for our model.

```python
corr_matrix = df_cleaned.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, cmap='viridis', annot=True, fmt = '.2f')
plt.title('Correlation Heatmap of All Features')
plt.show()
```

### Our correlation analysis has identified a cluster of features with strong positive correlations to the 'Pathologic' NSP class, with values ranging from 0.5 to over 0.7. This group notably includes ALTV , ASTV , and DP. This is a crucial insight, as it confirms that the primary indicators of fetal distress are abnormal heart rate variability and significant, prolonged decelerations. These features will form the core of our predictive model.
